```python
import pandas as pd
import random
import matplotlib.pyplot as plt
from collections import defaultdict
from rdkit import Chem

# Load the Excel file
file_path = "/kaggle/input/dataset20k/Data.xlsx"  # Update path if needed
df = pd.read_excel(file_path, sheet_name="Sheet1")

# Extract SMILES strings
smiles_list = df["SMILES_1"].dropna().astype(str).tolist()

# Function to build n-gram frequency dictionary
def build_ngram_counts(n):
    ngram_counts = defaultdict(int)
    for smiles in smiles_list:
        for i in range(len(smiles) - (n - 1)):
            ngram = smiles[i:i + n]
            ngram_counts[ngram] += 1
    return ngram_counts

# Function to convert counts to probabilities
def compute_ngram_probs(ngram_counts):
    total_ngrams = sum(ngram_counts.values())
    return {ng: count / total_ngrams for ng, count in ngram_counts.items()}

# Function to generate SMILES using n-gram probabilities
def generate_smiles_ngram(ngram_probs, n, length=15):
    if not ngram_probs:
        return None

    # Start with a random (n-1)-character sequence from known SMILES data
    start_seq = random.choice(list(ngram_probs.keys()))[:n-1]
    generated = list(start_seq)

    for _ in range(length - (n - 1)):
        last_n_minus_1 = "".join(generated[-(n - 1):])
        candidates = [ng[-1] for ng in ngram_probs.keys() if ng[:(n - 1)] == last_n_minus_1]
        probabilities = [ngram_probs[last_n_minus_1 + c] for c in candidates] if candidates else None

        if candidates and probabilities:
            next_char = random.choices(candidates, weights=probabilities, k=1)[0]
            generated.append(next_char)
        else:
            break  # Stop if no valid continuation

    return "".join(generated)

# Generate molecules for n values from 2 to 19 and validate them
valid_molecules_all = []
valid_counts = []
uniqueness_scores = []

for n in range(2, 20):
    ngram_probs = compute_ngram_probs(build_ngram_counts(n))
    generated_smiles = [generate_smiles_ngram(ngram_probs, n) for _ in range(1000)]

    # Validate molecules
    valid_smiles = [s for s in generated_smiles if s and Chem.MolFromSmiles(s) is not None]
    unique_valid_smiles = set(valid_smiles)

    valid_counts.append(len(valid_smiles))
    uniqueness = len(set(generated_smiles)) / len(generated_smiles) if generated_smiles else 0
    uniqueness_scores.append(uniqueness)

    valid_molecules_all.extend(valid_smiles)

    print(f"{n}-gram model: Generated {len(valid_smiles)} valid molecules out of 1000 "
          f"({(len(valid_smiles)/1000):.2%} valid), Uniqueness: {uniqueness:.2%}")

# Plot graph: Number of Valid Molecules vs N
plt.figure(figsize=(12, 5))

# Subplot 1: Valid molecules
plt.subplot(1, 2, 1)
plt.plot(range(2, 20), valid_counts, marker='o', linestyle='-', color='blue', label="Valid Molecules")
plt.title("Number of Valid Molecules vs N-gram Size")
plt.xlabel("N-gram Size (n)")
plt.ylabel("Number of Valid Molecules")
plt.grid(True, linestyle="--", alpha=0.7)
plt.xticks(range(2, 20))
plt.legend()

# Subplot 2: Uniqueness
plt.subplot(1, 2, 2)
plt.plot(range(2, 20), uniqueness_scores, marker='s', linestyle='-', color='green', label="Uniqueness")
plt.title("Uniqueness of Generated SMILES vs N-gram Size")
plt.xlabel("N-gram Size (n)")
plt.ylabel("Uniqueness (Ratio)")
plt.grid(True, linestyle="--", alpha=0.7)
plt.xticks(range(2, 20))
plt.legend()

plt.tight_layout()
plt.show()

print(f"\nTotal valid molecules generated across all models: {len(valid_molecules_all)}")
```

```python
import pandas as pd
import random
import matplotlib.pyplot as plt
from collections import defaultdict
from rdkit import Chem

# Load the Excel file
file_path = "/kaggle/input/dataset20k/Data.xlsx"  # Update path if needed
df = pd.read_excel(file_path, sheet_name="Sheet1")

# Extract SMILES strings
smiles_list = df["SMILES_1"].dropna().astype(str).tolist()

# Function to build n-gram frequency dictionary
def build_ngram_counts(n):
    ngram_counts = defaultdict(int)
    for smiles in smiles_list:
        for i in range(len(smiles) - (n - 1)):
            ngram = smiles[i:i + n]
            ngram_counts[ngram] += 1
    return ngram_counts

# Function to convert counts to probabilities
def compute_ngram_probs(ngram_counts):
    total_ngrams = sum(ngram_counts.values())
    return {ng: count / total_ngrams for ng, count in ngram_counts.items()}

# Function to generate SMILES using n-gram probabilities
def generate_smiles_ngram(ngram_probs, n, length=15):
    if not ngram_probs:
        return None

    start_seq = random.choice(list(ngram_probs.keys()))[:n-1]
    generated = list(start_seq)

    for _ in range(length - (n - 1)):
        last_n_minus_1 = "".join(generated[-(n - 1):])
        candidates = [ng[-1] for ng in ngram_probs.keys() if ng[:(n - 1)] == last_n_minus_1]
        probabilities = [ngram_probs[last_n_minus_1 + c] for c in candidates] if candidates else None

        if candidates and probabilities:
            next_char = random.choices(candidates, weights=probabilities, k=1)[0]
            generated.append(next_char)
        else:
            break
    return "".join(generated)

# Generation settings
n_values = range(2, 21)  # n from 2 to 20
gen_counts = [1000, 2000, 3000, 4000, 5000]  # number of molecules

# Storage
validity_results = {gen: [] for gen in gen_counts}
uniqueness_results = {gen: [] for gen in gen_counts}

# Main loop
for gen_n in gen_counts:
    print(f"\n==== Generating {gen_n} molecules ====")
    for n in n_values:
        ngram_probs = compute_ngram_probs(build_ngram_counts(n))
        generated_smiles = [generate_smiles_ngram(ngram_probs, n) for _ in range(gen_n)]

        valid_smiles = [s for s in generated_smiles if s and Chem.MolFromSmiles(s) is not None]
        unique_smiles = set(generated_smiles)

        validity = len(valid_smiles) / gen_n
        uniqueness = len(unique_smiles) / gen_n

        validity_results[gen_n].append(validity)
        uniqueness_results[gen_n].append(uniqueness)

        print(f"n={n:2d} | Valid: {len(valid_smiles)} ({validity:.2%}), "
              f"Unique: {len(unique_smiles)} ({uniqueness:.2%})")

# Plotting all graphs
fig, axes = plt.subplots(5, 2, figsize=(14, 20))
fig.suptitle("Validity and Uniqueness vs N-gram Size", fontsize=18)

for idx, gen_n in enumerate(gen_counts):
    row = idx
    axes[row][0].plot(n_values, validity_results[gen_n], marker='o', color='blue')
    axes[row][0].set_title(f"Validity (Generated: {gen_n})")
    axes[row][0].set_xlabel("N-gram Size (n)")
    axes[row][0].set_ylabel("Validity")
    axes[row][0].grid(True)

    axes[row][1].plot(n_values, uniqueness_results[gen_n], marker='s', color='green')
    axes[row][1].set_title(f"Uniqueness (Generated: {gen_n})")
    axes[row][1].set_xlabel("N-gram Size (n)")
    axes[row][1].set_ylabel("Uniqueness")
    axes[row][1].grid(True)

plt.tight_layout(rect=[0, 0.03, 1, 0.97])
plt.show()
```

```python
import seaborn as sns
import numpy as np

# Convert to DataFrames
valid_df = pd.DataFrame(validity_results, index=n_values)
uniq_df = pd.DataFrame(uniqueness_results, index=n_values)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.heatmap(valid_df, annot=True, fmt=".2f", cmap="Blues")
plt.title("Validity Heatmap")
plt.xlabel("Number of Molecules")
plt.ylabel("N-gram Size")

plt.subplot(1, 2, 2)
sns.heatmap(uniq_df, annot=True, fmt=".2f", cmap="Greens")
plt.title("Uniqueness Heatmap")
plt.xlabel("Number of Molecules")
plt.ylabel("N-gram Size")

plt.tight_layout()
plt.show()
```
